```python
movies = df_movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.sample()
```

```python
movies.dropna(inplace = True)
```

```python
movies.duplicated().sum()
```

```python
movies.isnull().sum()
```

```python
import ast
```

```python
def convert(text):
    list = []
    for i in ast.literal_eval(text):
        list.append(i['name'])
    return list

# CHANGING FORMAT FROM STRING TO LIST BY USING AST.LITERAL_EVAL(TEXT)
```

```python
movies.genres = movies['genres'].apply(convert)
```

```python
movies.keywords = movies['keywords'].apply(convert)
```

```python
movies.sample()
```

```python
movies.cast = movies['cast'].apply(convert)
movies.cast = movies['cast'].apply(lambda x : x[0:3])
```

```python
movies.sample()
```

```python
movies['cast'][0]
```

```python
def fetch_director(text):
    list = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            list.append(i['name'])
    
    return list

```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.sample()
```

```python
def collapse(text):
    l1 = []
    for i in text:
        l1.append(i.replace(" " , ""))

    return l1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)

# the same function can be written as movies['cast'] = movies['cast'].apply[ lambda x : (i.replace(" ","") for i in x]
```

```python
movies.sample(2)
```

```python
movies['overview'] = movies['overview'].apply(lambda x : x.split())
```

```python
movies.sample(2)
```

```python
movies['tags'] = movies['overview'] +  movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df = movies.drop(columns = ['overview',	'genres',	'keywords',	'cast',	'crew'])
```

```python
new_df.sample(3)
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x : " ".join(x))
```

```python
new_df.head()
```

```python
cv = CountVectorizer(max_features = 5000, stop_words='english')
```

```python
cv
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)

# we were facing the similar words having different feature column issue like 'actor' and 'actors' are similar words but they were treated different so we stem them up
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags']
```

```python
print(cv.fit_transform(new_df['tags']).toarray().shape)
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
# as we can see we have 4806 rows (movies) and 5000 words in each row
```

```python
# now we will calculate cosine distance to find out similarity of each movie with others 
# eventually we are directly using similarity function as it will be 1 stage solution
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
# we can see we got 4806*4806 table because we are calculating distance of each vetor with other vectors present in the dataset
```

```python
# now we need to fetch top 5 similar vectors and recommend their name as list
```

```python
sorted(similarity[0] , reverse = True)
```

```python
# we can get the sorted vectors but we lost which vector are these top vectors as we lost their index(column index)
```

```python
# so we will enumerate and make it index similarity (key-value) pair and then convert it to list
```

```python
list(enumerate(similarity))
```

```python
# we will get an array of array and we need to sort then in descending order
```

```python
# sorted(list(enumerate(similarity[5])) , reverse = True , key = lambda x : x[1])
```

```python
def recommend(movie):
    movie = "".join(movie.lower().split())
    title_list = new_df['title'].apply(lambda x: "".join((x.lower().split())))
    
    if movie in title_list.values:
        idx = title_list[title_list == movie].index[0]
        similar_movie = sorted(list(enumerate(similarity[idx])) , reverse = True , key = lambda x : x[1])
    
        for i in similar_movie[1:6]:
            print(new_df.iloc[i[0]].title.title())
            
    # .title() is used to print in 'Batman & Robin' this format because we change it to 'batman and robin' format
    
    else :
        print('Please enter another movie name')
```

**Try recommendation**

```python
recommend('Titanic')
```
